# Streaming and Batch

## Streaming

Output is given as and when chunks are produced

```python
model.stream("Why parrots speak")
```

```python
for chunk in model.stream("Why parrots speak"):
    print(chunk.text, end= "", flush= True)
```

```python
for chunk in model.stream("Why parrots speak"):
    print(chunk.text, end= "|", flush= True)
```

## Batch

Multiple imputs processed parallelly

```python
response = model.batch([
    "Why parrots speak",
    "Why airoplanes fly",
    "Write a poem"
])

for res in response:
    print(res)
```

```python
response = model.batch([
    "Why parrots speak",
    "Why airoplanes fly",
    "Write a poem",
],
config = {"max_concurrency": 5 }
)

for res in response:
    print(res.text,end = "\n\n\n")
```
